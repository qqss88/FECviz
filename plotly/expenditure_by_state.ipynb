```python
exp_df = pd.read_csv('oppexp.txt', sep='|', dtype=str, header=None)


exp_df = exp_df[[10, 11, 12,13]]
exp_df
```

```python
exp_df = exp_df[[10, 13]]
exp_df.columns = ['state', 'expditure']
exp_df.head()
```

```python
counts = exp_df.groupby(exp_df.state).size().reset_index(name='counts').sort_values('counts')
counts
```

```python
counts = counts[counts.counts > 5000]
counts.shape
```

```python
# counts
# counts = counts[-6:]
# counts
counts.state = counts.state.str.strip()
counts
```

```python
counts['text'] = counts['state'] + '<br>' + 'total:'+ counts['counts'].astype(str)
counts    
```

```python
import plotly.plotly as py

scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
      [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = counts['state'],
        z = counts['counts'].astype(float),
        locationmode = 'USA-states',
        text = counts['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "expenditure counts")
        ) ]

layout = dict(
        title = '2017-2018 expenditure counts by State<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```
