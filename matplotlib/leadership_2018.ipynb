```python
%time

from elasticsearch import helpers, Elasticsearch
import csv

# connect
es = Elasticsearch([{'host': 'localhost', 'port': 9200}])

# prepare index
idx_name = "ofec_leadership"
if es.indices.exists(idx_name):
    es.indices.delete(index=idx_name)
es.indices.create(index=idx_name, ignore=400)

# inject data
with open('leadership2018.csv') as f:
    reader = csv.DictReader(f)
    helpers.bulk(es, reader, index=idx_name, doc_type='data')
```
